```python
import pandas as pd
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
# Display dataset info nicely
print("Dataset Loaded Successfully!")
print(f"Shape of the data: {df.shape[0]} rows √ó {df.shape[1]} columns")
print("\n Preview of the first 5 rows:\n")

display(df.head())  # better than print(df.head()) in Jupyter
```

```python
df.info()
```

```python
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
if 'date_added' in df.columns:
    df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype(str).str.strip()
    print("\n‚úÖ Data cleaned successfully!")
```

```python
print("üìã Columns in dataset:", df.columns.tolist())

# Count unique values in important columns
for col in ['type', 'country', 'rating']:
    if col in df.columns:
        print(f"\nUnique values in {col}: {df[col].nunique()}")
```

```python
# Count of movies vs TV shows
if 'type' in df.columns:
    print("\nüé• Count of Movies vs TV Shows:")
    print(df['type'].value_counts())
```

```python
# Top 10 countries
if 'country' in df.columns:
    print("\nüåç Top 10 Countries by Number of Titles:")
    print(df['country'].value_counts().head(10))
```

```python
# Top genres
if 'listed_in' in df.columns:
    genres = df['listed_in'].str.split(',').explode().str.strip()
    print("\nüé≠ Top 10 Genres:")
    print(genres.value_counts().head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```python
if 'type' in df.columns:
    plt.figure(figsize=(6,4))
    sns.countplot(x='type', data=df, palette='pastel')
    plt.title('Movies vs TV Shows on Netflix')
    plt.show()
```

```python
# 2. Release Year Distribution
if 'release_year' in df.columns:
    plt.figure(figsize=(10,5))
    sns.histplot(df['release_year'], bins=30, kde=True, color='skyblue')
    plt.title('Distribution of Release Year')
    plt.xlabel('Release Year')
    plt.show()
```

```python
# --- Top Genres on Netflix ---
import matplotlib.pyplot as plt
import seaborn as sns

if 'listed_in' in df.columns:
    # Split the comma-separated genres into individual ones
    genres = df['listed_in'].dropna().str.split(',').explode().str.strip()

    # Get top 10 most common genres
    top_genres = genres.value_counts().head(10)

    # Plot
    plt.figure(figsize=(10, 5))
    sns.barplot(x=top_genres.values, y=top_genres.index, palette='viridis')

    # Titles and labels
    plt.title('üé≠ Top 10 Genres on Netflix', fontsize=14, fontweight='bold')
    plt.xlabel('Number of Titles', fontsize=12)
    plt.ylabel('Genre', fontsize=12)

    # Display count values on the bars
    for index, value in enumerate(top_genres.values):
        plt.text(value + 5, index, str(value), va='center', fontsize=10)

    plt.tight_layout()
    plt.show()
else:
    print("‚ö†Ô∏è Column 'listed_in' not found in dataset!")
```
